```python
result = []
for i in range(2, 10):
    for j in range(1, 10):
        cur_result = f'{i} x {j} = {i * j}'
        result.append(cur_result)

loop_count = 0
for item in result:
    if loop_count % 9 == 0:
        print(f'== {loop_count // 9 + 2}단 ==')
    print(item)
    loop_count += 1
```

```python
for index, item in enumerate(result):
    if index % 9 == 0:
        print()
    print(item)
```
