```python
N, A, output_lst = list(map(int, input().split())), list(), list()

for i in range(N[0]):
    A.insert(0, int(input()))    

for i, money in enumerate(A):
    while money <= N[1]:
        div = divmod(N[1], money)
        output_lst.append(div[0])
        N[1] = div[1]
print(sum(output_lst))
```
