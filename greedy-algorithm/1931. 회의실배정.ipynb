```python
N = int(input())
Is = list()

for n in range(N):
    I = tuple(map(int, input().split()))
    Is.append(I)

Is = sorted(Is)

len_lst = list()

for n in range(N-1):
    lst = list()
    idx = n
    current_i = Is[idx]
    
    while idx < N-1:
        if current_i[1] <= Is[idx+1][0]:
            current_i = Is[idx+1]
            idx = Is.index(current_i)+1
            lst.append(current_i)
            continue
        else:
            idx += 1
            continue
    
    len_lst.append(len(lst))
print(max(len_lst)+1)
```
