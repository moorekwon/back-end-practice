```python
change = 1000 - int(input())
monetary_units = [500, 100, 50, 10, 5, 1]
cnt = list()
for unit in monetary_units:
    div = divmod(change, unit)
    
    if div[1] == 0:
        cnt.append(div[0])
        break
        
    elif div[0] == 0:
        change = div[1]
        
    else:
        cnt.append(div[0])
        change = div[1]
        
print(sum(cnt))
```

```python
change = 1000 - int(input())
monetary_units = [500, 100, 50, 10, 5]
cnt = 0
for unit in monetary_units:
    while change >= unit:
        change -= unit
        cnt += 1
cnt += change
print(cnt)
```
